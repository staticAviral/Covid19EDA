```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

# Importing main dataset

```python
df=pd.read_csv('covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
df.head()
```

# Keeping only required columns

```python
df=df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
```

# Renaming column names

```python
df.columns=['Date','State','Cured','Deaths','Confirmed']
```

# Looking at the earlier dates

```python
df.head()
```

# Looking at the latest dates

```python
df.tail()
```

# Looking at the particular date - (2021-07-04)

```python
date=df[df.Date=='2021-07-04']
```

```python
date.head()
```

# Sorting data w.r.t. number of confirmed cases

```python
max_confirmed_cases=date.sort_values(by='Confirmed',ascending=False)[['Date','State','Confirmed']]
```

```python
max_confirmed_cases
```

# Getting states with maximum number of confirmed cases

```python
top_states_confirmed=max_confirmed_cases.head()
```

```python
top_states_confirmed
```

# Making bar plot for states with top confirmed cases

```python
sns.barplot(x='State',y='Confirmed',data=top_states_confirmed,hue='State')
plt.show()
```

# Sorting data w.r.t. number of death cases

```python
max_death_cases=date.sort_values(by='Deaths',ascending=False)[['Date','State','Deaths']]
max_death_cases
```

# Top 5 states with maximum death cases

```python
top_states_death=max_death_cases.head()
```

```python
top_states_death
```

# Making a bar plot for states with maximum death cases

```python
sns.barplot(x='State',y='Deaths',data=top_states_death,hue='State')
plt.show()
```

# Sorting data w.r.t number of cured cases

```python
max_cured_cases=date.sort_values(by='Cured',ascending=False)[['Date','State','Cured']]
max_cured_cases
```

# Top 5 states with maximum cured cases

```python
top_states_cured=max_cured_cases.head()
```

```python
top_states_cured
```

# Making a bar plot for the top 5 states with maximum cured cases

```python
sns.barplot(x='State',y='Cured',hue='State',data=top_states_cured)
plt.show()
```
